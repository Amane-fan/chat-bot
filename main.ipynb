```python
import streamlit as st
import pandas as pd

# 页面标题
st.title("我的第一个 Streamlit 应用 🚀")

# 侧边栏
with st.sidebar:
    name = st.text_input("你的名字", "访客")
    age = st.slider("年龄", 0, 100, 25)

# 主内容区
st.write(f"你好，{name}！你今年 {age} 岁。")

# 数据展示
data = pd.DataFrame({
    '月份': ['1月', '2月', '3月'],
    '销售额': [100, 150, 200]
})
st.line_chart(data.set_index('月份'))

# 文件上传
uploaded_file = st.file_uploader("上传 CSV 文件")
if uploaded_file:
    df = pd.read_csv(uploaded_file)
    st.dataframe(df)
```
